```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[18]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True,axis=1)
```

```python
df_tracks.columns
```

```python
df_tracks.duration
```

```python
corr_df = df_tracks.drop(['key','mode','explicit'],axis=1).corr(method="pearson")
plt.figure(figsize=(15,5))
heatmap = sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="PuRd",linewidth=1,linecolor="White")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y="loudness",x="energy",color="c",line_kws={"color":"orange"}).set(title="Loudness Vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df,y="popularity",x="acousticness",color="orange",line_kws={"color":"green"}).set(title="Popularity Vs Acousticness Correlation")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,color='orange',kind="hist").set(title="Number of songs per year")
```

```python
sns.distplot(years,kde=False,bins=120,color="orange")
plt.title("Number of songs per year")
plt.xlabel("Year", fontsize=13)
plt.ylabel("Songs Count", fontsize=13)

import warnings
warnings.filterwarnings("ignore")
```

```python
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(y=total_dr, x=years ,ax=ax,errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax=plt.subplots(figsize=fig_dims)
fig = sns.lineplot(y=total_dr, x=years,ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
df_genre = pd.read_csv("C:/D Drive/DataSets/SpotifyFeatures.csv")
df_genre.head()
```

```python
df_genre['duration_s']=df_genre['duration_ms'].apply(lambda x:round(x/1000))
```

```python
plt.title("Duration of Songs in different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y="genre",x="duration_ms",data=df_genre)
plt.xlabel("Duration in milli-seconds")
plt.ylabel("Genres")
```

```python
plt.title("Duration of Songs in different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y="genre",x="duration_s",data=df_genre)
plt.xlabel("Duration in seconds")
plt.ylabel("Genres")
```

```python
plt.title("Duration of Songs in different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.lineplot(x="genre",y="duration_ms",data=df_genre)
plt.xlabel("Genres")
plt.ylabel("Duration in milli-seconds")
plt.xticks(rotation=90)
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y="genre",x="popularity",data=famous).set(title="Top 5 Genres by Popularity")
plt.xlabel("Popularity")
plt.ylabel("Genre")
```
